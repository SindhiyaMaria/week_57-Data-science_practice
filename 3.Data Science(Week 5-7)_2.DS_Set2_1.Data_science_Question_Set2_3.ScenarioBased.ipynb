 Scenario 1: Flight Delay Analysis

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns


# Sample Flight Delay Data (20 flights in minutes)

np.random.seed(42)  

delays = [12, 8, 45, 3, 67, 15, 22, 5, 89, 34, 
          7, 18, 55, 2, 120, 28, 41, 9, 76, 14]

df = pd.DataFrame({'Flight_Delay_Minutes': delays})

print("=== Flight Delay Analysis Report ===\n")
print(df.describe().round(2))
```

```python

# 1. Percentiles

print("\n=== Percentiles ===")
percentiles = [25, 50, 75, 90, 95, 99]
for p in percentiles:
    value = np.percentile(delays, p)
    print(f"{p}th Percentile: {value:.2f} minutes")
```

```python

# 2. Outlier Detection (IQR Method)

Q1 = np.percentile(delays, 25)
Q3 = np.percentile(delays, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in delays if x < lower_bound or x > upper_bound]

print("\n=== Outlier Detection ===")
print(f"Q1 (25th): {Q1:.2f}")
print(f"Q3 (75th): {Q3:.2f}")
print(f"IQR: {IQR:.2f}")
print(f"Lower Bound: {lower_bound:.2f}")
print(f"Upper Bound: {upper_bound:.2f}")
print(f"Detected Outliers: {outliers}")
```

```python

# 3. Distribution Analysis

print("\n=== Distribution Analysis ===")
print(f"Mean Delay     : {np.mean(delays):.2f} minutes")
print(f"Median Delay   : {np.median(delays):.2f} minutes")
print(f"Std Deviation  : {np.std(delays, ddof=1):.2f} minutes")
print(f"Skewness       : {stats.skew(delays):.4f}")   # Positive skew = right tail
print(f"Kurtosis       : {stats.kurtosis(delays):.4f}")
```

```python

# Visualization (Save as images)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(delays, kde=True, bins=10, color='skyblue')
plt.title('Flight Delay Distribution')
plt.xlabel('Delay (minutes)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(y=delays, color='lightgreen')
plt.title('Box Plot - Flight Delays with Outliers')
plt.ylabel('Delay (minutes)')

plt.tight_layout()
plt.savefig('flight_delay_analysis.png')
plt.show()

print("\n Analysis completed! Graph saved as 'flight_delay_analysis.png'")
```

Scenario 2:  Employee Salary Analysis

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns


# Sample Employee Salary Data (in INR LPA)

np.random.seed(42)

salaries = [4.2, 5.8, 6.5, 3.9, 12.5, 7.2, 4.8, 15.0, 5.5, 8.1,
            6.8, 4.5, 9.3, 5.2, 22.0, 7.8, 6.1, 4.1, 10.5, 5.9,
            8.7, 4.9, 13.2, 6.3, 5.1]

df = pd.DataFrame({'Salary_LPA': salaries})

print("=== Employee Salary Analysis Report ===\n")
print(df.describe().round(2))
```

```python

# 1. Central Tendency

print("\n=== Central Tendency ===")
print(f"Mean Salary     : ₹{np.mean(salaries):.2f} LPA")
print(f"Median Salary   : ₹{np.median(salaries):.2f} LPA")
print(f"Mode Salary     : ₹{stats.mode(salaries).mode:.2f} LPA")
```

```python

# 2. Skewness Analysis

skewness = stats.skew(salaries)
print("\n=== Skewness Analysis ===")
print(f"Skewness Value  : {skewness:.4f}")

if skewness > 0.5:
    print("→ The salary distribution is **Positively Skewed** (Right Skewed)")
    print("  Interpretation: Most employees earn below average, few high earners pull the mean up.")
elif skewness < -0.5:
    print("→ The salary distribution is **Negatively Skewed** (Left Skewed)")
else:
    print("→ The salary distribution is approximately **Symmetric**")
```

```python

# 3. Percentiles

print("\n=== Key Percentiles ===")
for p in [25, 50, 75, 90, 95]:
    value = np.percentile(salaries, p)
    print(f"{p}th Percentile: ₹{value:.2f} LPA")
```

```python

# 4. Outlier Detection (IQR Method)

Q1 = np.percentile(salaries, 25)
Q3 = np.percentile(salaries, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in salaries if x < lower_bound or x > upper_bound]

print("\n=== Outlier Detection ===")
print(f"Q1 (25th)     : ₹{Q1:.2f} LPA")
print(f"Q3 (75th)     : ₹{Q3:.2f} LPA")
print(f"IQR           : ₹{IQR:.2f} LPA")
print(f"Upper Bound   : ₹{upper_bound:.2f} LPA")
print(f"Detected Outliers: {outliers}")
```

```python

# Visualizations

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(salaries, kde=True, bins=12, color='royalblue')
plt.title('Employee Salary Distribution')
plt.xlabel('Salary (LPA)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(y=salaries, color='lightcoral')
plt.title('Salary Box Plot - Outliers Visible')
plt.ylabel('Salary (LPA)')

plt.tight_layout()
plt.savefig('employee_salary_analysis.png')
plt.show()

print("\n Analysis completed! Graph saved as 'employee_salary_analysis.png'")
```

Scenario 3: Product Sales Analysis

Scenario 4: Student Exam Performance Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Sample Data: Marks out of 100
np.random.seed(42)
students = pd.DataFrame({
    'Mathematics': [78, 85, 65, 92, 71, 88, 76, 82, 69, 95, 74, 81],
    'Science':     [82, 79, 71, 88, 75, 91, 68, 84, 73, 89, 77, 80],
    'English':     [75, 88, 69, 85, 72, 90, 78, 83, 70, 92, 76, 79]
})

print("=== Student Exam Performance Analysis ===\n")
print(students.describe().round(2))

# Subject-wise Analysis
print("\n=== Subject-wise Statistics ===")
for subject in students.columns:
    print(f"\n{subject}:")
    print(f"  Mean     : {students[subject].mean():.2f}")
    print(f"  Median   : {students[subject].median():.2f}")
    print(f"  Std Dev  : {students[subject].std():.2f}")
    print(f"  Pass %   : {(students[subject] >= 40).mean()*100:.1f}%")

# Correlation Analysis
print("\n=== Correlation Matrix ===")
print(students.corr().round(3))

# Visualizations
plt.figure(figsize=(12, 5))
sns.boxplot(data=students)
plt.title('Performance Comparison Across Subjects')
plt.ylabel('Marks')
plt.show()

sns.pairplot(students)
plt.suptitle('Relationship Between Subjects', y=1.02)
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data: Marks out of 100
np.random.seed(42)
students = pd.DataFrame({
    'Student_ID': range(1, 31),
    'Mathematics': np.random.normal(72, 12, 30).round(1),
    'Science': np.random.normal(68, 15, 30).round(1),
    'English': np.random.normal(75, 10, 30).round(1)
})

print("=== Student Exam Performance Analysis ===\n")
print(students.describe().round(2))

# Subject-wise Analysis
subjects = ['Mathematics', 'Science', 'English']
print("\n=== Subject-wise Statistics ===")
for sub in subjects:
    print(f"\n{sub}:")
    print(f"  Mean     : {students[sub].mean():.2f}")
    print(f"  Median   : {students[sub].median():.2f}")
    print(f"  Std Dev  : {students[sub].std():.2f}")
    print(f"  Skewness : {stats.skew(students[sub]):.3f}")

# Correlation Analysis
print("\n=== Correlation Matrix ===")
print(students[subjects].corr().round(3))

# Visualizations
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.boxplot(data=students[subjects])
plt.title('Subject-wise Performance')

plt.subplot(1, 3, 2)
sns.heatmap(students[subjects].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')

plt.subplot(1, 3, 3)
sns.pairplot(students[subjects])
plt.suptitle('Pairwise Relationships', y=1.02)

plt.tight_layout()
plt.savefig('student_performance_analysis.png')
plt.show()
```

Scenario 5: Clinical Trial for Diabetes Medication


```python
from scipy import stats
import numpy as np
import pandas as pd

# Sample Data: Blood Sugar Level after 30 days (mg/dL)
np.random.seed(42)
medication_group = [135, 128, 142, 119, 131, 125, 138, 122, 129, 134]
placebo_group    = [158, 165, 172, 149, 160, 155, 168, 175, 152, 161]

# Hypothesis Test
t_stat, p_value = stats.ttest_ind(medication_group, placebo_group, equal_var=False)

print("=== Clinical Trial - Hypothesis Testing ===\n")
print(f"Medication Group Mean : {np.mean(medication_group):.2f} mg/dL")
print(f"Placebo Group Mean    : {np.mean(placebo_group):.2f} mg/dL")
print(f"T-Statistic           : {t_stat:.4f}")
print(f"P-Value               : {p_value:.6f}")

if p_value < 0.05:
    print("\n Result: Medication is **statistically significant** in reducing blood sugar.")
    print("   We reject the null hypothesis.")
else:
    print("\n Result: No significant difference found.")

# Effect Size (Cohen's d)
def cohens_d(x, y):
    nx, ny = len(x), len(y)
    dof = nx + ny - 2
    return (np.mean(x) - np.mean(y)) / np.sqrt(((nx-1)*np.std(x, ddof=1)**2 + (ny-1)*np.std(y, ddof=1)**2) / dof)

print(f"Cohen's d (Effect Size): {cohens_d(medication_group, placebo_group):.3f} (Large effect)")
```

```python
import numpy as np
from scipy import stats
import pandas as pd

# Sample Clinical Trial Data (Blood Sugar Level after treatment)
np.random.seed(42)
medication_group = np.random.normal(110, 12, 25).round(1)   # Lower is better
placebo_group = np.random.normal(135, 18, 25).round(1)

print("=== Clinical Trial - Hypothesis Testing ===\n")

# Descriptive Statistics
print("Medication Group:")
print(pd.Series(medication_group).describe().round(2))
print("\nPlacebo Group:")
print(pd.Series(placebo_group).describe().round(2))

# Independent T-Test
t_stat, p_value = stats.ttest_ind(medication_group, placebo_group, equal_var=False)

print("\n=== Hypothesis Test Results ===")
print(f"T-Statistic     : {t_stat:.4f}")
print(f"P-Value         : {p_value:.6f}")

alpha = 0.05
if p_value < alpha:
    print(f"\n Result: Reject Null Hypothesis (p < {alpha})")
    print("   The medication is **statistically effective** in reducing blood sugar levels.")
else:
    print("\n Result: Fail to Reject Null Hypothesis")
    print("   No strong evidence that the medication is effective.")

# Effect Size (Cohen's d)
def cohens_d(group1, group2):
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    pooled_std = np.sqrt(((n1-1)*var1 + (n2-1)*var2) / (n1 + n2 - 2))
    return (np.mean(group1) - np.mean(group2)) / pooled_std

effect_size = cohens_d(medication_group, placebo_group)
print(f"Effect Size (Cohen's d): {effect_size:.3f} (Medium to Large effect)")
```
