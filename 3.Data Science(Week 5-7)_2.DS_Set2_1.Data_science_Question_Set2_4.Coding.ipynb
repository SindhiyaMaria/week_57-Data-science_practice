Question 1: Flight Delay Analysis

```python
import numpy as np

delays = [15, 30, 45, 20, 25, 100, 5, 60, 35, 50,
          120, 80, 10, 75, 90, 200, 55, 40, 70, 150]

# Percentiles
percentiles = np.percentile(delays, [10, 25, 50, 75, 90])
print("Percentiles (10th, 25th, 50th, 75th, 90th):", np.round(percentiles, 2))

# IQR & Outliers
Q1 = np.percentile(delays, 25)
Q3 = np.percentile(delays, 75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = [x for x in delays if x < lower or x > upper]

print("IQR:", round(IQR, 2))
print("Outliers:", outliers)
```

Question 2: Employee Salary Analysis

```python
import numpy as np
from scipy import stats

salaries = [30000, 32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000, 55000,
            60000, 62000, 65000, 67000, 70000, 72000, 75000, 80000, 85000, 90000]

mean_salary = np.mean(salaries)
median_salary = np.median(salaries)
mode_salary = stats.mode(salaries).mode

print("Mean Salary:", mean_salary)
print("Median Salary:", median_salary)
print("Mode Salary:", mode_salary)

if mean_salary < median_salary:
    print("The data is Left Skewed (Negative Skew)")
else:
    print("The data is Right Skewed (Positive Skew)")
```

Question 3: Student Exam Performance Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset
data = {
    'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Mathematics': [85, 78, 92, 88, 70, 95, 60, 80, 90, 76],
    'Science': [80, 85, 88, 70, 75, 92, 55, 82, 89, 78],
    'English': [78, 74, 85, 80, 68, 90, 50, 77, 83, 72]
}

# Create DataFrame
df = pd.DataFrame(data)

# -------------------------------
# Descriptive Statistics
# -------------------------------
print("Descriptive Statistics")
print(df.describe())

# -------------------------------
# Histogram
# -------------------------------
plt.figure(figsize=(10, 5))

plt.hist(df['Mathematics'], bins=5, alpha=0.5, label='Mathematics')
plt.hist(df['Science'], bins=5, alpha=0.5, label='Science')
plt.hist(df['English'], bins=5, alpha=0.5, label='English')

plt.title("Histogram of Student Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# -------------------------------
# Correlation Analysis
# -------------------------------
correlation = df[['Mathematics', 'Science', 'English']].corr()

print("\nCorrelation Matrix")
print(correlation)

# Correlation Graph
plt.figure(figsize=(6, 4))

sns.heatmap(correlation, annot=True, cmap='coolwarm')

plt.title("Correlation Analysis")
plt.show()

# -------------------------------
# HeatMap
# -------------------------------
plt.figure(figsize=(8, 5))

sns.heatmap(df[['Mathematics', 'Science', 'English']],
            annot=True,
            cmap='YlGnBu')

plt.title("HeatMap of Student Scores")
plt.show()
```

Question 4: Clinical Trial Hypothesis Test

```python
from scipy import stats
import numpy as np

medication_group = [110, 115, 108, 102, 107, 99, 111, 104, 109, 101]
placebo_group = [120, 125, 130, 122, 128, 119, 124, 127, 123, 126]

t_stat, p_value = stats.ttest_ind(medication_group, placebo_group)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis: The medication is effective.")
else:
    print("Fail to reject the null hypothesis.")
```
